```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "output0.1.csv"  # Reemplaza con la ruta de tu archivo
df = pd.read_csv(file_path, delimiter='\s+', header=0)

# Muestra las primeras filas del DataFrame
df
```

```python
# Extrae la primera columna (suponiendo que se llama "Jet_Eta")
jet_eta_column = df.iloc[:, 0]

# Calcula el nÃºmero de entradas, la media y la desviaciÃ³n estÃ¡ndar
num_entries = len(jet_eta_column)
mean_value = jet_eta_column.mean()
std_deviation = jet_eta_column.std()

fig, ax = plt.subplots(figsize=(10, 6))

# Grafica el histograma de la primera columna
plt.hist(jet_eta_column, bins=5000, edgecolor='blue', color='white', linewidth=1.2)  # PersonalizaciÃ³n del contorno y relleno
plt.xlabel('Jet_Eta', fontsize=14)  # Ajusta el tamaÃ±o de la fuente del eje x
plt.ylabel('Frecuencia', fontsize=14)  # Ajusta el tamaÃ±o de la fuente del eje y
plt.title('Histograma de Jet_Eta', fontsize=16)  # Ajusta el tamaÃ±o de la fuente del tÃ­tulo

# Agrega informaciÃ³n al grÃ¡fico con mÃ¡s espacio
plt.text(0.02, 0.92, f'Entradas: {num_entries}', transform=ax.transAxes, fontsize=12)
plt.text(0.02, 0.88, f'Media: {mean_value:.6f}', transform=ax.transAxes, fontsize=12)
plt.text(0.02, 0.84, f'Des Est: {std_deviation:.6f}', transform=ax.transAxes, fontsize=12)

plt.tight_layout()  # Ajusta automÃ¡ticamente el espaciado para evitar superposiciones

plt.show()
```
