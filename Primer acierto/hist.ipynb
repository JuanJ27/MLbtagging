```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = "outputEta_Phi.csv"  # Reemplaza con la ruta de tu archivo
df = pd.read_csv(file_path, delimiter='\s+', header=0)

# Muestra las primeras filas del DataFrame
df
```

```python
# Extrae la primera columna (suponiendo que se llama "Jet_Eta")
jet_eta_column = df.iloc[:, 0]

# Crear la figura y el eje
fig, ax = plt.subplots()

# Histograma
n, bins, patches = ax.hist(jet_eta_column, bins=500, edgecolor='blue', color='white')

# Ajustar el estilo
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# TÃ­tulo y etiquetas
ax.set_title('Jet.Eta')
ax.set_xlabel('Jet.Eta')

# Mostrar informaciÃ³n en la parte superior derecha
entries = len(jet_eta_column)
mean = np.mean(jet_eta_column)
std_dev = np.std(jet_eta_column)

textstr = '\n'.join((
    r'Entries %d' % (entries, ),
    r'Mean %.6f' % (mean, ),
    r'Std Dev %.3f' % (std_dev, )))

# Propiedades del texto
props = dict(boxstyle='round', facecolor='white', alpha=0.5)

# PosiciÃ³n del texto
ax.text(0.75, 0.95, textstr, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)

plt.show()
```
